```python
import numpy as np
import matplotlib.pyplot as plt
array = np.random.randint(low=-10, high=11, size=(10, 3))
fig, ax = plt.subplots()
ax.scatter(array[:,0], array[:,1], c=array[:,2], cmap='viridis')
ax.set_xlabel('Column 1')
ax.set_ylabel('Column 2')
ax.set_title('Random Integers in a (10, 3) Numpy Array')
plt.show()
```
